## pi, sin, cos, exp

```python
x = np.linspace(0,10*np.pi,100)
print 'x'

sin = np.sin(x)
print 'sin(x)'

cos = np.cos(x)
print 'cos(x)'

exp = np.exp(x/10)
print 'e^x'
```

## example plots - regular plot

```python
print plt.style.available
plt.style.use('fivethirtyeight')
```

```python
plt.figure(figsize=(10,8))

plt.plot(x,x, color='r', label='y = x', linewidth=5, alpha=.5)
plt.scatter(x[::10],x[::10], c='b', s=x*1000)

plt.title('Example Plot', fontsize=24, color='orange')
plt.xlabel('x axis', fontsize=16)
plt.ylabel('y axis', fontsize=16)

plt.grid(True,linewidth=1)
plt.legend()
plt.show()
```

## example plots - subplots

```python
f, ax = plt.subplots(1,4,figsize=(24,6))

ax[0].plot(x,sin, color='b', linewidth=3)
ax[1].plot(x,cos, color='g', linewidth=1)
ax[2].plot(x,exp, color='c', linewidth=1)
ax[3].plot(x, sin * exp)

ax[0].set_title(r'$sin(x)$',fontsize=24)
ax[1].set_title(r'$cos(x)$',fontsize=24)
ax[2].set_title(r'$e^{x}$',fontsize=24)
ax[3].set_title(r'$sin(x)*e^{x}$',fontsize=24)

ax[0].grid(True,c='r',alpha=.2)
ax[1].grid(True,c='b',alpha=.2)
ax[2].grid(True,c='g',alpha=.2)
ax[3].grid(True,c='y',alpha=.2)
```

## example plots - playing with color and markers

```python
# print dir(cm) # prints all colors
f, ax = plt.subplots(1,4,sharey=True,figsize=(24,6))

norm = mpl.colors.Normalize(vmin=min(x),vmax=max(x))

ax[0].plot(x,sin*exp, linewidth=4, alpha=.1)
ax[1].plot(x,sin*exp, linewidth=4, alpha=.1)
ax[2].plot(x,sin*exp, linewidth=4, alpha=.1)
ax[3].plot(x,sin*exp, linewidth=4, alpha=.1)

cmap0 = cm.viridis
m0 = cm.ScalarMappable(norm=norm, cmap=cmap0)
ax[0].scatter(x,sin*exp, c=m0.to_rgba(x), marker='*', s=10*x)

cmap1 = cm.winter
m1 = cm.ScalarMappable(norm=norm, cmap=cmap1)
ax[1].scatter(x,sin*exp, c=m1.to_rgba(x), marker='o', s=10*x)

cmap2 = cm.rainbow
m2 = cm.ScalarMappable(norm=norm, cmap=cmap2)
ax[2].scatter(x,sin*exp, c=m2.to_rgba(x), marker='x', s=10*x)

cmap3 = cm.plasma
m3 = cm.ScalarMappable(norm=norm, cmap=cmap3)
ax[3].scatter(x,sin*exp, c=m3.to_rgba(x), marker='^', s=10*x)
```

### more compact way to make previous plots

```python
f, ax = plt.subplots(1,4, sharey=True, figsize=(24,6))

colors = [cm.viridis,cm.winter,cm.rainbow,cm.plasma]
markers = ['*','o','x','^']

for idx, item in enumerate(zip(colors,markers)):
    c, mk = item
    m = cm.ScalarMappable(norm=norm, cmap=c)
    ax[idx].scatter(x,sin*exp, c=m.to_rgba(x), marker=mk, s=x*10)
    ax[idx].plot(x,sin*exp, linewidth=4, alpha=.1)
```

## using random numbers to generate fictional dataset -> histograms

```python
x = np.random.randn(10000)
y = np.random.randn(10000)

plt.scatter(x,y,s=1)
numBins = 20
```

```python
plt.hist(x,bins=numBins,alpha=.4)
plt.show()

binCount, binEdges = np.histogram(x,bins=numBins)

indexSmallestBin = np.argmin(binCount)
indexLargestBin = np.argmax(binCount)

print binCount
print binEdges
print ''
print 'which bin, index of bin, amount in bin, middle value of bin'
print 'smallest',indexSmallestBin, min(binCount), (binEdges[indexSmallestBin] + binEdges[indexSmallestBin + 1])/2.
print 'largest',indexLargestBin, max(binCount), (binEdges[indexLargestBin] + binEdges[indexLargestBin + 1])/2.

```

## 2D histograms and contour plots

```python
plt.hist2d(x,y,bins=numBins)
plt.colorbar()
plt.show()
```

```python
xx = np.linspace(min(x),max(x),numBins)
yy = np.linspace(min(y),max(y),numBins)

X, Y = np.meshgrid(xx, yy)
H, binXedges, binYedges = np.histogram2d(x, y, bins=numBins)
plt.contourf(X,Y,H)
plt.colorbar()
```
